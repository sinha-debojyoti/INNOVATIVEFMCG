```python
%config IPCompleter.greedy=True
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data_BISCUITS.csv")
df
```

```python
df["CREAM"]=df["CREAM"].astype('category')
```

```python
df.info()
```

```python
df.describe()
```

```python
dummies = pd.get_dummies(df["CREAM"])
dummies
```

```python
df = pd.concat([df, dummies], axis="columns")
df
```

```python
df = df.drop(["CREAM"], axis="columns")
df
```

```python
X = df.drop("SELLING_PRICE", axis="columns")
X
```

```python
y = df["SELLING_PRICE"]
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.predict(X)
```

```python
model.score(X,y)
```

```python
# MRP, WEIGHT, Cream NO, Cream YES
model.predict([[60,150,0, 1]])
```

```python
import pickle
```
